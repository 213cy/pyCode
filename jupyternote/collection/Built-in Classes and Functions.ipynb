```python
b = vars(__builtins__)
c =  [ k for k in b.items() if  not hasattr( k[1] ,'__cause__' )  ]
d = dict(  k for k in c if  isinstance(k[1],type) or type(k[1])==type(abs)  )
for k in d.items() :print( '%-30s %s' % k ) 
print(len(d))
# d.pop('__build_class__')
# d.pop('__import__')
# d.pop('breakpoint')
```

```python
# from inspect import signature
# from inspect import ismodule
# from inspect import isclass
# from inspect import ismethod
# from inspect import isfunction
from inspect import *

for f,g in d.items():
    print(isclass(g),end='\t')
    try:
        print(f,str(signature(g)) )
    except:
        t = g.__doc__
        print('*',end='')
        print(t[0:t.find('\n')])
```

```python
from inspect import signature
a=0
r=[]
b = vars(__builtins__)
for f,g in b.items():
#     print('\t\t',f,g)
#     if isclass(g) and issubclass(g,BaseException):
# https://docs.python.org/3/library/exceptions.html#exception-hierarchy
    if  hasattr( g ,'__cause__' ) :
        continue
    elif type(g)!=type and type(g)!=type(abs):
#         print('xxxxx', f, g)
        continue
    else: 
        a += 1
        try:
            t = str(signature(g))
            r.append(t)
            print(f, t)
        except ValueError:
            t = g.__doc__
            r.append(t)
            t = t[t.find('('):t.find('\n')]
            print(f,'*',t[t.find('('):t.find('\n')] )
print(a,len(r))
```
